Size of different application are different but they are object data type. We can easily convert this column to float. I did the same by the function below.

```python
def manage_size(items):

  if str(items)[-1] == 'M':
    res = items.replace("M", "")
    return res
  elif str(items)[-1] == 'K':
    res2 = items.replace("K", "")
    res2 = res2/1024
    return res2
  else:
    return str(np.NaN)
```

```python
df1['Size'] = df1['Size'].apply(manage_size)
```

```python
df1['Size'] = df1.Size.astype(float)
```
